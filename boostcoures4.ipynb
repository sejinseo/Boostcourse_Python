```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/boostcourse/diabetes.csv'
df = pd.read_csv(file_path, encoding='cp949')
df.shape
```

```python
df.head()
```

```python
split_count = int(df.shape[0] * 0.8)
split_count
```

```python
train = df[:split_count].copy()
train.shape
```

```python
test = df[split_count:].copy()
test.shape
```

```python
feature_name = train.columns[:-1].tolist()
feature_name
```

```python
label_name = train.columns[-1]
label_name
```

```python
X_train = train[feature_name]
print(X_train.shape)
X_train.head()
```

```python
y_train = train[label_name]
print(y_train.shape)
y_train.head()
```

```python
X_test = test[feature_name]
print(X_test.shape)
X_test.head()
```

```python
y_test = test[label_name]
print(y_test.shape)
y_test.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
model
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
y_predict[:5]
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 20))
tree = plot_tree(model,
          feature_names=feature_name,
          filled=True, fontsize=10)
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_tree = export_graphviz(model, feature_names = feature_name,
filled=True)
graphviz.Source(dot_tree)
```

```python
model.feature_importances_
```

```python
sns.barplot(x=model.feature_importances_, y=feature_name)
```

```python
diff_count = abs(y_test - y_predict).sum()
diff_count
```

```python
(len(y_test) - diff_count) / len(y_test) * 100
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict) * 100
```

```python
model.score(X_test, y_test) * 100
```
